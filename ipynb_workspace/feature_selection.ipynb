# max-min selector
0. feature 들을 normalize 해서 sphere 위에 분포하도록 함
1. feature vector 하나를 고정
2. feature vector 와 거리가 가장 먼 다른 feature vector 를 선택 (거리는 sphere 위의 arc length 로 설정)
3. 순차적으로 이미 선택된 feature vector 들과 거리가 가장 먼 feature vector 들을 선택해나감

```python
def maxminSelector(data, n, initial):
    data = (data-data.mean())/data.std()
    features = data.columns
    dissmat = pd.DataFrame(np.arccos(np.corrcoef(data.values.T)),index=features,columns=features)
    result = [initial]
    n-=1
    while(n!=0):
        n -= 1
        result += [dissmat[result].min(axis=1).idxmax()]
    return result
    
```

# clustering selector
1. feature vector 들 간의 거리를 이용해 single linkage clustering 을 적용
2. 각 cluster 마다 feature vector 하나씩 선택

```python
def myslink(dist):
    N = len(dist)
    p = np.zeros(N, dtype = int)
    l = np.zeros(N, dtype = float)
    m = np.zeros(N, dtype = float)
    
    for n in np.arange(N):
        # S1
        p[n] = n
        l[n] = np.inf
        
        # S2
        m[:n] = dist[:n,n]
        
        # S3
        for i in np.arange(n):
            if l[i] >= m[i]:
                m[p[i]] = np.min((m[p[i]],l[i]))
                l[i] = m[i]
                p[i] = n
            else:
                m[p[i]] = np.min((m[p[i]],m[i]))
                
        # S4
        for i in np.arange(n):
            if l[i] >= l[p[i]]:
                p[i] = n
                
    return p,l


def myclustering(points, epsilon,metrictype='arcmetric'):
    n = 30
    if metrictype == 'arcmetric':
        dist = np.arccos(np.corrcoef(points.values.T))
    elif metrictype == 'Euclidean':
        dist = np.zeros((30,30))
        for i in np.arange(30):
            for j in np.arange(i):
                dist[i,j] = np.linalg.norm(points.iloc[:,i]-points.iloc[:,j])
        dist = dist + dist.T
    else:
        dist = np.zeros(30)
        
    pointer,levelset = myslink(dist)
    contained = np.where(levelset < epsilon)
    ptindex = [i for i in range(n)]
    result = []
    num = 0
    subsets = []
    while len(ptindex) != 0:
        tempcluster = []
        temp = ptindex.pop(0)
        while(1):
            tempcluster = tempcluster + [list(points.columns)[temp]]
            if not(temp in contained[0]):
                break
            temp = pointer[temp]
        subsets.append(tempcluster)
        
    tempind = list(set([c[-1] for c in subsets]))
    for i,endnum in zip(range(len(tempind)),tempind):
        cluster_ = []
        for c in subsets:
            if c[-1] == endnum:
                cluster_.extend(c)
                cluster_ = list(set(cluster_))
        result.append(cluster_)
    return result
```

```python
def get_cluster_output(mds_f):
    r=0
    n_clust = np.inf
    clust = []
    clusters = {}
    limit = 0
    while(n_clust!=1 and limit<1000):
        mds_f_ = pd.DataFrame(mds_f.T,columns=features)
        temp = myclustering(mds_f_,0.002*r,'Euclidean')
        limit += 1
        r += 1
        if clust!=temp:
            n_clust = len(temp)
            clusters[n_clust] = temp
            clust = temp
    return clusters

temp= get_cluster_output(mds_f)
print(temp)
```
