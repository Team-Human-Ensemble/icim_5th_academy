```python
def maxminSelector(data, n, initial):
    data = (data-data.mean())/data.std()
    dissmat = pd.DataFrame(np.arccos(np.corrcoef(data.values.T)),index=data.columns,columns=data.columns)
    result = [initial]
    n-=1
    while(n!=0):
        n -= 1
        result += [dissmat[result].min(axis=1).idxmax()]
    return result

```
